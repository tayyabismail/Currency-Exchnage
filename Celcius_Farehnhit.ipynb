```python
import requests
import ipywidgets as widgets
from IPython.display import display

# ExchangeRate-API URL
api_url = "https://v6.exchangerate-api.com/v6/d10cdf4e0852ad5e6a789660/latest/USD"

# Function to get exchange rates
def get_exchange_rates(api_url):
    response = requests.get(api_url)
    data = response.json()
    if response.status_code == 200:
        return data['conversion_rates']
    else:
        print("Error fetching data from the API. Please check the API key or URL.")
        return None

# Fetch exchange rates
rates = get_exchange_rates(api_url)

# Create dropdowns for currency selection and input for amount
if rates:
    currencies = list(rates.keys())
    from_currency_dropdown = widgets.Dropdown(options=currencies, description='From:')
    to_currency_dropdown = widgets.Dropdown(options=currencies, description='To:')
    amount_input = widgets.FloatText(description='Amount:', value=1.0)

    # Create a button to perform conversion
    convert_button = widgets.Button(description='Convert')
    output_label = widgets.Label()

    # Function to perform conversion
    def convert_currency(b):
        from_currency = from_currency_dropdown.value
        to_currency = to_currency_dropdown.value
        amount = amount_input.value

        if from_currency and to_currency:
            conversion_rate = rates[to_currency] / rates[from_currency]
            converted_amount = amount * conversion_rate
            output_label.value = f"{amount} {from_currency} is equal to {converted_amount:.2f} {to_currency}"
        else:
            output_label.value = "Please select both currencies for conversion."

    # Bind the function to the button click event
    convert_button.on_click(convert_currency)

    # Display the widgets
    display(from_currency_dropdown, to_currency_dropdown, amount_input, convert_button, output_label)
else:
    print("Could not fetch exchange rates. Please try again later.")
```
